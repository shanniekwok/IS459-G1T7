<b>Import libaries</b>

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

<b>Import CSV files</b>

```python
df = pd.read_csv('informations_households.csv')
```

```python
# Assuming your data is in a pandas DataFrame called df
# Grouping the data by 'Acorn' and counting the number of 'LCLid' in each category
acorn_counts = df.groupby('Acorn')['LCLid'].count().reset_index()

# Create a larger figure
plt.figure(figsize=(20, 10))

# Create a barplot with color coding based on Acorn type
sns.barplot(x='Acorn', y='LCLid', data=acorn_counts, palette='Set2')

# Add labels to each bar
for index, row in acorn_counts.iterrows():
    plt.text(index, row['LCLid'] + 0.2, str(row['LCLid']), ha='center', color='black')

# Adding labels and title
plt.xlabel('Acorn Category')
plt.ylabel('Number of LCLid')
plt.title('Number of LCLid in Each Acorn Category')

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming your data is in a pandas DataFrame called df
# Grouping the data by 'Acorn' and counting the number of 'LCLid' in each category
acorn_counts = df.groupby('Acorn_grouped')['LCLid'].count().reset_index()

# Create a larger figure
plt.figure(figsize=(20, 10))

# Create a barplot with color coding based on Acorn type
sns.barplot(x='Acorn_grouped', y='LCLid', data=acorn_counts, palette='Set2')

# Add labels to each bar
for index, row in acorn_counts.iterrows():
    plt.text(index, row['LCLid'] + 0.2, str(row['LCLid']), ha='center', color='black')

# Adding labels and title
plt.xlabel('Acorn_grouped Category')
plt.ylabel('Number of LCLid')
plt.title('Number of LCLid in Each Acorn Category')

# Show the plot
plt.show()
```

```python
# Create a new DataFrame to count 'Std' and 'ToU' for each 'Acorn' type
acorn_counts = pd.crosstab(df['Acorn'], df['stdorToU'])

# Plot the stacked bar chart
ax = acorn_counts.plot(kind='bar', stacked=True, figsize=(20, 10), color=['skyblue', 'orange'])

# Adding labels to each segment of the stacked bars
for p in ax.patches:
    height = p.get_height()
    width = p.get_width()
    x_position = p.get_x() + width / 2  # Position to center the label on the bar
    y_position = p.get_y() + height / 2  # Position to place the label vertically in the middle
    if height > 0:  # Only add labels to non-zero heights
        ax.text(x_position, y_position, str(int(height)), ha='center', va='center', color='black')

# Adding labels and title
plt.xlabel('Acorn Category')
plt.ylabel('Count of LCLid')
plt.title('Stacked Bar Chart of Std and ToU in Each Acorn Category')

# Adding legend to indicate the categories
plt.legend(title='Type', labels=['Std', 'ToU'])

# Show the plot
plt.show()
```
